```python
import matplotlib.pyplot as plt
import numpy as np

# Define your data
x_values = np.array([1, 2, 3, 4])  # X-values where you want your data points to be spread
y_values = np.array([8.8300e-1,1.0384,8.0800e-1,7.1766e-1])  # Y-values for your data points
errors = np.array([4.46e-2,8.21e-2,9.48e-2,1.48e-1])  # Vertical error bars for each data point

# Create the figure and axis
fig, ax = plt.subplots()

# Create the histogram bars (you can customize the width)
width = 0.4  # Adjust the width of the bars as needed
plt.scatter(x_values, y_values)
plt.errorbar(x_values, y_values, yerr=errors, fmt="o")

# Customize the plot
ax.set_xlabel('p_T (GeV)')
ax.set_ylabel('Scale Factor')
ax.set_title('2016 Scale Factor Measurement')

# Set the x-axis ticks and labels
ax.set_xticks(x_values)
ax.set_xticklabels(['[350, 450)', '[450, 500)', '[500, 600)', '[600, inf)'])  # Replace with your desired labels

# Add a legend
ax.legend()

ax.grid()

# Display the plot
plt.show()
```
