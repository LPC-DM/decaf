```python
import matplotlib.pyplot as plt
import numpy as np
import mplhep as hep
import os
plt.style.use(hep.style.CMS)

year='2018'

# Define your data
x_values = np.array([0.5, 1.5, 2.5])  # X-values where you want your data points to be spread
y_values = np.array([0.82, 0.75, 0.81])  # Y-values for your data points
errors = np.array([np.sqrt(0.07**2 + 0.11**2), np.sqrt(0.06**2 + 0.06**2), np.sqrt(0.05**2 + 0.01**2)])  # Vertical error bars for each data point

# Create the figure and axis
fig, ax = plt.subplots()

# Create the histogram bars (you can customize the width)
width = 0.4  # Adjust the width of the bars as needed
plt.scatter(x_values, y_values)
plt.errorbar(x_values, y_values, yerr=errors, fmt="o")

# Customize the plot
ax.set_xlabel(r'AK15 Jet $p_T$ (GeV)')
ax.set_ylabel('Scale Factor')
ax.set_ylim(0,1.2)

# Set the x-axis ticks and labels
ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(['350', '400', '500', '2500'])  # Replace with your desired labels

# Add a legend
ax.legend()

ax.grid()

hep.cms.text(ax=ax, loc=0, text='',fontsize=25)
hep.cms.lumitext(ax=ax, text=year+" (13 TeV)",fontsize=25) 

# Display the plot
plt.show()

os.system('mkdir -p ../plots/doublebsf/results/'+year+'/')
plot_path = os.path.abspath('../plots/doublebsf/results/'+year+'/')
plot_name = year+'_sf.png'
fig.savefig(os.path.join(plot_path, plot_name))
```
